## Question #1
What top 10 states in the US (United States) reported highest rates of human trafficking in 2021? 

```python
import pandas as pd
```

```python
topstates = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Final Project/2021_Human Trafficking Rates-Grid view (1).csv')
print(topstates)
```

```python
topstates.head(10)
```

```python
sorted_topstates = topstates.sort_values('Cases identified', ascending=False)
print(sorted_topstates)
```

```python
sorted_topstates.head(10)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
heatmap_data = topstates[['State', 'Year', 'Signals received', 'Cases identified']]
heatmap_table = heatmap_data.pivot(index='State', columns='Year', values='Signals received')
sns.set(font_scale=0.8)
fig, ax = plt.subplots(figsize=(12, 10))
sns.heatmap(heatmap_table, annot=True, fmt='g', cmap='YlGnBu', cbar_kws={'label': 'Signals Received'}, ax=ax)
ax.set_title('Signals Received by State and Year')
ax.set_xlabel('Year')
ax.set_ylabel('State')
plt.show()
```

## Question #2
What was the most common type of trafficking reported in the 5 top trafficking states in 2021? 

```python
types = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Final Project/Common Types of Trafficking-Grid view (1).csv')
print(types)
```

```python
types.head()
```

```python
import matplotlib.pyplot as plt
```

```python
selected_states = ['California', 'Texas', 'Florida', 'New York', 'Michigan']
selected_columns = ['Illicit Massage/Spa Business', 'Hotel/Motel-Based']
filtered_types = types[types['State'].isin(selected_states)][['State'] + selected_columns]
filtered_types.set_index('State', inplace=True)
colors = ['purple', 'green']
filtered_types.plot(kind='bar', color=colors)
plt.xlabel('State')
plt.ylabel('Number of Cases')
plt.title('Most Common Type of Trafficking by Top States')
plt.legend(title='Trafficking Type')
plt.show()
```

## Question #3
What was the most common way of contacting the human trafficking hotline in 2021 vs. in 2014? 

```python
form = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Final Project/Form of Contacting the Hotline-Grid view (1).csv')
print(form)
```

```python
form.head(10)
```

```python
form_2021 = form[form['Year of Reports'] == 2021]
form_2014 = form[form['Year of Reports'] == 2014]
grouped_2021 = form_2021.groupby('Signal Types')['Number of Reports'].sum()
grouped_2014 = form_2014.groupby('Signal Types')['Number of Reports'].sum()
most_common_2021 = grouped_2021.idxmax()
most_common_2014 = grouped_2014.idxmax()
print("Most common reporting method in 2021:", most_common_2021)
print("Most common reporting method in 2014:", most_common_2014)
```

## Question #4
What was the most common way individuals contacted the National Human Trafficking Hotline from 2014 to 2021? 

```python
import matplotlib.pyplot as plt
```

```python
gdata = form.groupby(['Signal Types', 'Year of Reports'])['Number of Reports'].sum()
r_data = gdata.unstack(level='Signal Types')
r_data.plot(kind='line', marker='o')
plt.xlabel('Year')
plt.ylabel('Number of Reports')
plt.title('Human Trafficking Reports by Signal Type')
plt.legend(title='Signal Types', bbox_to_anchor=(1, 0.5), loc='center left')
plt.subplots_adjust(right=0.85) 
plt.show()
```

## Question #7
What is the mean ratio of adults to minors across the top 10 human trafficking states in 2021? 

```python
dem_rates = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Final Project/Demographics of Hotline-Grid view.csv')
print(dem_rates)
```

```python
dem_rates.head(10)
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
df_plot = dem_rates.drop(columns=['Name'])
fig, ax = plt.subplots()
_, _ = ax.pie(df_plot.iloc[0], labels=df_plot.columns, colors=['blue', 'green', 'red', 'yellow', 'purple', 'orange'])
plt.title('Overview of Demographic Totals')
totals = df_plot.iloc[0].astype(int).sum()
categories = df_plot.columns
percentages = (df_plot.iloc[0].astype(int) / totals) * 100
annotations = [f'{category}: {percentage:.2f}%' for category, percentage in zip(categories, percentages)]
ax.set_aspect('equal')
legend_labels = [f'{label} ({percentage:.2f}%)' for label, percentage in zip(categories, percentages)]
legend_handles = [mpatches.Patch(color='blue'), mpatches.Patch(color='green'), \
                  mpatches.Patch(color='red'), mpatches.Patch(color='yellow'), \
                  mpatches.Patch(color='purple'), mpatches.Patch(color='orange')]
plt.legend(legend_handles, legend_labels, loc='center left', \
           bbox_to_anchor=(1.5, 0.5), fontsize=10)
plt.show()
```

```python
dem_rates['Adult-to-Minor Ratio'] = dem_rates['Adult'] / dem_rates['Minor']
average_ratio = dem_rates['Adult-to-Minor Ratio'].mean()
print("Average Adult-to-Minor Ratio:", average_ratio)
```
